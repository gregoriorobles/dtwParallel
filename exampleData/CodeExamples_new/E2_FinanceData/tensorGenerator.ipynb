## Code to process information from 505 financial companies

```python
# importing the required modules
import glob
import pandas as pd
import numpy as np
  
# specifying the path to csv files
path = "../../Data/E2_FinanceData/all_data"
  
# csv files in the path
files = glob.glob(path + "/*.CSV")

  
# defining an empty list to store 
# content
data_frame = pd.DataFrame()
content = []
  
# checking all the csv files in the specified path
for filename in files:
    # reading content of csv file
    df = pd.read_csv(filename, sep=";", index_col=None)
    content.append(df)
  
# converting content to data frame
data_frame = pd.concat(content)
```

```python
valueMax = np.inf
ifinal = 0
for i in range(len(content)):
    if valueMax > content[i].shape[0]:
        valueMax = content[i].shape[0]
        ifinal = i

valueMax = 500
df = pd.DataFrame()
content_f = []
for i in range(len(content)):
    content_f.append(content[i][0:valueMax])
df = pd.concat(content_f)
print(df.shape)
df_aux = df[['Symbol', 'Timestamp']].reset_index(drop=True)
df = df[['Adjclose', 'Close', 'High', 'Low', 'Open', 'Volume']]
```

```python
import pandas as pd
from sklearn import preprocessing

x = df.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df_aux_2 = pd.DataFrame(x_scaled, columns=df.columns)
df = df_aux_2.join(df_aux)
print(df.shape)
```

```python
def dataframe_to_tensor(df, eliminateColumn, columns, valueMax):       
    eliminateColumn = True
    columns=["Timestamp"]

    _, idx = np.unique(df.Symbol, return_index=True)
    listSamples = np.array(df.Symbol)[np.sort(idx)]

    index = df.index

    for i in range(len(listSamples)):
        df_trial = df[df.Symbol == listSamples[i]]
        
        if df_trial.shape[0] < valueMax:
            for new_values in range(int(valueMax-df_trial.shape[0])):
                new_data = pd.DataFrame(df_trial[-1:].values, columns=df_trial.columns)
                df_trial = df_trial.append(new_data)
            df_trial = df_trial.reset_index(drop=True)
            
        if eliminateColumn:
            df_trial = df_trial.drop(columns=columns)
        
        if i == 0:
            X = np.array(df_trial)
            X = X.reshape(1, valueMax, df.shape[1] - len(columns))
        else:
            X_2 = np.array(df_trial)
            X_2 = X_2.reshape(1, valueMax, df.shape[1] - len(columns))
            X = np.append(X, X_2, axis=0)

    X = np.delete(X, [-1], axis=2)

    return X
```

```python
X_data_tensor = dataframe_to_tensor(
                 df, 
                 eliminateColumn=True, columns=["Timestamp"], 
                 valueMax=valueMax)
print(X_data_tensor.shape)
np.save(
   "../../Data/E2_FinanceData/FinanceData_500days_norm.npy", X_data_tensor
)
```
