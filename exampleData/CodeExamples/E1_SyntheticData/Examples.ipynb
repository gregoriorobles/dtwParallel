### # E1

```python
from dtwParallel import dtw_functions as dtw
from scipy.spatial import distance 
import numpy as np

x = [2.5, 4.3, 6.6, 8.0, 1, 0, 0, 1, 5.5, 15.2]
y = [12.1, 0, 0, 1, 1, 6.4, 3.5, 1, 0, 0]

dtw_distance = dtw.dtw(x, y, dist=distance.euclidean, check_errors=True)
print("DTW distance: ", np.round(dtw_distance,4))

dtw_distance = dtw.dtw(x, y, dist=distance.canberra)
print("DTW distance: ", np.round(dtw_distance,4))

dtw_distance = dtw.dtw(x, y, dist=distance.euclidean, get_visualization=True)
print("DTW distance: ", np.round(dtw_distance,4))
```
